```python
import pandas as pd
from matplotlib import pyplot as plt
a=pd.read_csv("Movie_classification.csv")
a
```

```python
a.isnull().sum()
```

```python
a.dropna(inplace=True)
```

```python
a.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a['Budget']=le.fit_transform(a['Budget'])
a['Trailer_views']=le.fit_transform(a['Trailer_views'])
a['Avg_age_actors']=le.fit_transform(a['Avg_age_actors'])
a['Num_multiplex']=le.fit_transform(a['Num_multiplex'])
a['Genre']=le.fit_transform(a['Genre'])
a['Collection']=le.fit_transform(a['Collection'])
a['Twitter_hastags']=le.fit_transform(a['Twitter_hastags'])
a['3D_available']=le.fit_transform(a['3D_available'])
a
```

```python
x=a.iloc[:,0:18]
x
```

```python
y=a.iloc[:,18]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

## NAIVE BAYES CLASSIFICATION

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
p=nb.predict(x_test)
p
```

```python
from sklearn import metrics
nb_score=metrics.accuracy_score(y_test,p)
nb_score
```

## DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(random_state=0)
dtc.fit(x_train,y_train)
```

```python
p=dtc.predict(x_test)
p
```

```python
from sklearn import metrics
dtc_score=metrics.accuracy_score(y_test,p)
dtc_score
```

## RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=7,random_state=0)
rfc.fit(x_train,y_train)
```

```python
p=rfc.predict(x_test)
p
```

```python
from sklearn import metrics
rfc_score=metrics.accuracy_score(y_test,p)
rfc_score
```

## LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression(random_state=0)
lgr.fit(x_train,y_train)
```

```python
p=lgr.predict(x_test)
p
```

```python
from sklearn import metrics
lgr_score=metrics.accuracy_score(y_test,p)
lgr_score
```

## SVM CLASSIFIER

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
p=sv.predict(x_test)
p
```

```python
from sklearn import metrics
svm_score=metrics.accuracy_score(y_test,p)
svm_score
```

## knn-k-nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
a=KNeighborsClassifier(n_neighbors=5)
a.fit(x_train,y_train)
```

```python
ra=a.predict(x_test)
ra
```

```python
from sklearn import metrics
knn_score=metrics.accuracy_score(y_test,ra)
knn_score
```

```python
df=pd.DataFrame({
    "Models":["Naive Bayes Classification","Decision Tree Classification","Random Forest Classification","Logistic Regression","SVM Classifier","knn Classifier"],
    "Accuracy Score":[nb_score,dtc_score,rfc_score,lgr_score,svm_score,knn_score]
})

import seaborn as sns
sns.barplot(y="Models",x="Accuracy Score",data=df)
df.sort_values(by="Accuracy Score",ascending=False)
```
